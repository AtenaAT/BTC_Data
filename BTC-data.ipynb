```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
```

```python
# create a client obj
cg = CoinGeckoAPI()

# use func to req data
# in us$ for past 30 days
BTC_data = cg.get_coin_market_chart_by_id(id = 'bitcoin', vs_currency='usd',
                                         days = 30)
```

```python
# BTC_data #unix timestamp & price
```

```python
BTC_price_data = BTC_data['prices']
```

```python
# conver nested list to dataframe
df = pd.DataFrame(BTC_price_data, columns = ['Timestamp', 'Price'])
```

```python
df
```

```python
# make timestamp readable
df['Date'] = pd.to_datetime(df['Timestamp'], unit= 'ms')
```
